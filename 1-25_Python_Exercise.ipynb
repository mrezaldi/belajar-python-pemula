```python
# Latihan 006 – Luas Persegi Panjang

# Mengambil input panjang dari pengguna dan mengkonversinya ke tipe data float
panjang = float(input("Masukkan panjang: "))
# Mengambil input lebar dari pengguna dan mengkonversinya ke tipe data float
lebar = float(input("Masukkan lebar: "))
# Menghitung luas persegi panjang dengan mengalikan panjang dan lebar
luas = panjang * lebar
# Menampilkan hasil perhitungan luas persegi panjang
print("Luas persegi panjang adalah:", luas)
```

```python
# Latihan 007 – Konversi Celcius ke Fahrenheit

# Mengambil input suhu dalam Celcius dari pengguna dan mengkonversinya ke tipe data float
celcius = float(input("Masukkan suhu (°C): "))
# Menghitung konversi dari Celcius ke Fahrenheit menggunakan rumus: F = (C × 9/5) + 32
fahrenheit = (celcius * 9/5) + 32
# Menampilkan hasil konversi suhu ke Fahrenheit
print("Suhu dalam Fahrenheit:", fahrenheit)
```

```python
# Latihan 008 – Konversi Menit ke Jam

# Mengambil input dari pengguna berupa total menit dan mengkonversi ke integer
menit = int(input("Masukkan total menit: "))
# Menghitung jumlah jam dengan pembagian integer (floor division)
jam = menit // 60
# Menghitung sisa menit dengan operator modulo
sisa = menit % 60
# Menampilkan hasil konversi dengan format string
print(f"Hasilnya: {jam} jam dan {sisa} menit")
```

```python
# Latihan 009 – Kuadratkan Angka

# Mengambil input angka dari pengguna dan mengkonversinya ke tipe data integer
angka = int(input("Masukkan angka: "))
# Menampilkan hasil kuadrat dari angka yang dimasukkan (angka pangkat 2)
print("Kuadrat dari angka tersebut adalah:", angka ** 2)
```

```python
# Latihan 010 – Rata-Rata dari 3 Angka

# Meminta input dari pengguna berupa tiga angka yang dipisahkan spasi
# Menggunakan split() untuk memisahkan input menjadi list string
# Menggunakan map() untuk mengkonversi setiap string menjadi float
# Menyimpan ketiga angka ke dalam variabel a, b, dan c
a, b, c = map(float, input("Masukkan tiga angka dipisahkan spasi: ").split())

# Menghitung rata-rata dari ketiga angka dengan menjumlahkan lalu membagi dengan 3
rata2 = (a + b + c) / 3

# Menampilkan hasil rata-rata ke layar
print("Rata-rata dari ketiga angka tersebut adalah:", rata2)
```

```python
# Latihan 011 – Menentukan Bilangan Genap atau Ganjil

# Mengambil input dari pengguna dan mengkonversinya ke tipe data integer
angka = int(input("Masukkan angka: "))
# Memeriksa apakah angka habis dibagi 2 (genap)
if angka % 2 == 0:
    # Jika habis dibagi 2, maka angka tersebut genap
    print("Angka tersebut adalah GENAP.")
else:
    # Jika tidak habis dibagi 2, maka angka tersebut ganjil
    print("Angka tersebut adalah GANJIL.")
```

```python
# Latihan 012 – Positif, Negatif, atau Nol

# Mengkonversi input pengguna menjadi tipe data float
angka = float(input("Masukkan angka: "))
  
# Memeriksa apakah angka lebih besar dari 0
if angka > 0:
    print("Bilangan positif.")
# Memeriksa apakah angka lebih kecil dari 0
elif angka < 0:
    print("Bilangan negatif.")
# Jika angka tidak positif dan tidak negatif, maka angka adalah 0
else:
    print("Bilangan nol.")
```

```python
# Latihan 013 – Cek Umur Dewasa atau Anak

# Meminta input umur dari pengguna dan mengkonversinya ke tipe data integer
umur = int(input("Masukkan umurmu: "))
  
# Memeriksa apakah umur lebih dari atau sama dengan 18
if umur >= 18:
    # Jika umur >= 18, cetak pesan bahwa pengguna sudah dewasa
    print("Kamu sudah dewasa.")
else:
    # Jika umur < 18, cetak pesan bahwa pengguna masih anak-anak
    print("Kamu masih anak-anak.")
```

```python
# Latihan 014 – Bandingkan Dua Angka

# Mengambil input angka pertama dari pengguna dan mengkonversinya ke tipe data float
a = float(input("Masukkan angka pertama: "))
# Mengambil input angka kedua dari pengguna dan mengkonversinya ke tipe data float
b = float(input("Masukkan angka kedua: "))
 
# Memeriksa apakah angka pertama lebih besar dari angka kedua
if a > b:
    # Jika a > b, tampilkan pesan bahwa angka pertama lebih besar
    print(f"{a} lebih besar dari {b}")
# Memeriksa apakah angka pertama lebih kecil dari angka kedua
elif a < b:
    # Jika a < b, tampilkan pesan bahwa angka kedua lebih besar
    print(f"{b} lebih besar dari {a}")
# Jika kedua kondisi di atas tidak terpenuhi (artinya a = b)
else:
    # Tampilkan pesan bahwa kedua angka sama besar
    print("Kedua angka sama besar.")
```

```python
# Latihan 015 – Evaluasi Nilai Huruf (Grading)

# Meminta input nilai dari pengguna dan mengkonversinya ke tipe data integer
nilai = int(input("Masukkan nilai (0–100): "))
  
# Menentukan grade berdasarkan rentang nilai
if nilai >= 90:
    grade = "A"  # Nilai 90-100 mendapatkan grade A
elif nilai >= 80:
    grade = "B"  # Nilai 80-89 mendapatkan grade B
elif nilai >= 70:
    grade = "C"  # Nilai 70-79 mendapatkan grade C
elif nilai >= 60:
    grade = "D"  # Nilai 60-69 mendapatkan grade D
else:
    grade = "E"  # Nilai di bawah 60 mendapatkan grade E
  
# Menampilkan hasil grade yang diperoleh
print(f"Nilai hurufmu adalah: {grade}")
```

```python
# Latihan 017 – Jumlahkan Angka 1-100

# Initialize a variable to store the sum
total = 0
# Loop through numbers from 1 to 100 (inclusive)
for i in range(1, 101):
    # Add each number to the running total
    total += i
# Print the final sum
print("Jumlah total dari 1 sampai 100 adalah:", total)
```

```python
# Latihan 018 – Tabel Perkalian

# Mengambil input angka dari pengguna dan mengkonversinya ke tipe data integer
angka = int(input("Masukkan angka: "))
  
# Membuat perulangan dari 1 sampai 10 untuk membuat tabel perkalian
for i in range(1, 11):
    # Mencetak hasil perkalian dalam format "angka x i = hasil"
    print(f"{angka} x {i} = {angka * i}")
```

```python
# Latihan 019 – Hitung Jumlah Huruf Vokal

# Mengambil input dari pengguna dan mengubahnya menjadi huruf kecil
kata = input("Masukkan sebuah kata: ").lower()
# Mendefinisikan string yang berisi semua huruf vokal
vokal = "aiueo"
# Inisialisasi variabel untuk menghitung jumlah huruf vokal
jumlah = 0

# Melakukan iterasi untuk setiap karakter dalam kata
for huruf in kata:
    # Memeriksa apakah karakter saat ini adalah huruf vokal
    if huruf in vokal:
        # Jika ya, tambahkan penghitung
        jumlah += 1

# Menampilkan hasil perhitungan jumlah huruf vokal
print(f"Jumlah huruf vokal: {jumlah}")
```

```python
# Latihan 020 – Hitung Mundur

# Meminta input dari pengguna berupa angka untuk memulai hitungan mundur
# dan mengkonversinya dari string ke integer
angka = int(input("Mulai hitung dari: "))

# Loop while yang akan berjalan selama nilai angka lebih besar atau sama dengan 0
while angka >= 0:
    # Menampilkan nilai angka saat ini
    print(angka)
    # Mengurangi nilai angka sebesar 1 pada setiap iterasi
    angka -= 1

# Menampilkan pesan setelah hitungan mundur selesai
print("Selesai!")
```

```python
# Latihan 022 – Tambahkan Buah ke Dalam Daftar

# Inisialisasi list buah dengan beberapa buah
buah = ["apel", "pisang", "jeruk"]
# Meminta input dari pengguna untuk menambahkan buah baru
baru = input("Masukkan buah baru: ")
# Menambahkan buah baru ke dalam list buah menggunakan metode append()
buah.append(baru)
# Menampilkan list buah yang sudah diperbarui
print("Daftar buah setelah ditambahkan:", buah)
```

```python
# Latihan 023 – Hapus Buah dari Daftar

# Membuat daftar buah
buah = ["apel", "pisang", "jeruk", "mangga"]
# Meminta input dari pengguna tentang buah yang ingin dihapus
hapus = input("Buah apa yang ingin dihapus? ")

# Memeriksa apakah buah yang ingin dihapus ada dalam daftar
if hapus in buah:
    # Jika ada, hapus buah tersebut dari daftar
    buah.remove(hapus)
    # Tampilkan daftar buah setelah penghapusan
    print("Buah telah dihapus:", buah)
else:
    # Jika buah tidak ditemukan dalam daftar, tampilkan pesan error
    print("Buah tidak ditemukan.")
```

```python
# Latihan 024 – Urutkan Daftar Buah Secara Alfabet

# Membuat daftar buah
buah = ["mangga", "apel", "pisang", "melon", "jeruk"]
# Mengurutkan daftar buah secara alfabetis menggunakan metode sort()
buah.sort()
# Menampilkan daftar buah yang sudah diurutkan
print("Buah setelah diurutkan:", buah)
```

```python
# Latihan 025 – Tampilkan Setiap Buah dengan Nomor

# Define a list of fruits
buah = ["apel", "pisang", "jeruk", "melon", "mangga"]

# Loop through the list with enumeration starting from 1
# enumerate() returns pairs of index and value
for i, item in enumerate(buah, 1):
    # Print each fruit with its number using f-string formatting
    print(f"{i}. {item}")
```
