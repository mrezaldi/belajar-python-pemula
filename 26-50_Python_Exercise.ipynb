```python
# Latihan 028 â€“ Cek Palindrom

# Meminta input dari pengguna dan mengubahnya menjadi huruf kecil
kata = input("Masukkan kata: ").lower()
# Memeriksa apakah kata tersebut sama jika dibaca dari belakang (palindrom)
# kata[::-1] adalah cara Python untuk membalik string
if kata == kata[::-1]:
    # Jika kondisi terpenuhi, kata tersebut adalah palindrom
    print("Kata ini adalah palindrom.")
else:
    # Jika kondisi tidak terpenuhi, kata tersebut bukan palindrom
    print("Bukan palindrom.")
```

```python
# Latihan 039 â€“ Fungsi Faktorial

def faktorial(n):
    hasil = 1
    for i in range(1, n + 1):
        hasil *= i
    return hasil

angka = int(input("Masukkan angka: "))
print(f"Faktorial dari {angka} adalah {faktorial(angka)}")
```

```python
# Latihan 041 â€“ Konverter Suhu Interaktif

pilih = input("Ketik 'C' untuk ke Fahrenheit atau 'F' ke Celcius: ").upper()
suhu = float(input("Masukkan suhu: "))
 
if pilih == "C":
    hasil = (suhu * 9/5) + 32
    print(f"Hasil: {hasil:.2f}Â°F")
elif pilih == "F":
    hasil = (suhu - 32) * 5/9
    print(f"Hasil: {hasil:.2f}Â°C")
else:
    print("Pilihan tidak valid.")
```

```python
# Latihan 042 â€“ Kalkulator Mini

a = float(input("Masukkan angka pertama: "))
op = input("Masukkan operator (+, -, *, /): ")
b = float(input("Masukkan angka kedua: "))
  
if op == "+":
    print("Hasil:", a + b)
elif op == "-":
    print("Hasil:", a - b)
elif op == "*":
    print("Hasil:", a * b)
elif op == "/":
    if b != 0:
        print("Hasil:", a / b)
    else:
        print("âš ï¸ Tidak bisa dibagi nol!")
else:
    print("Operator tidak valid.")
```

```python
# Latihan 043 â€“ Hitung Frekuensi Kata dalam Kalimat

kalimat = input("Masukkan kalimat: ").lower()
kata_list = kalimat.split()
frekuensi = {}
  
for kata in kata_list:
    frekuensi[kata] = frekuensi.get(kata, 0) + 1

print("Frekuensi kata:")
for k, v in frekuensi.items():
    print(f"{k}: {v}")
```

```python
# Latihan 044 â€“ Daftar Tugas (To-Do List)

tugas = []
  
while True:
    aksi = input("\nTambah / Hapus / Lihat / Keluar: ").lower()
 
    if aksi == "tambah":
        item = input("Tugas baru: ")
        tugas.append(item)
    elif aksi == "hapus":
        item = input("Tugas yang ingin dihapus: ")
        if item in tugas:
            tugas.remove(item)
    elif aksi == "lihat":
        print("Daftar tugas:", tugas)
    elif aksi == "keluar":
        break
    else:
        print("Perintah tidak dikenal.")
```

```python
# Latihan 045 â€“ Kalkulator Diskon Belanja

print("Selamat datang di Toko Python!")
total = float(input("Masukkan total belanja: "))

if total >= 500_000:
    diskon = 0.10
elif total >= 300_000:
    diskon = 0.05
else:
   diskon = 0.0
 
potongan = total * diskon
total_bayar = total - potongan

print(f"\nTotal Belanja : Rp {total:,.2f}")
print(f"Diskon        : Rp {potongan:,.2f}")
print(f"Total Bayar   : Rp {total_bayar:,.2f}")
```

```python
# Latihan 047 â€“ Cek Body Mass Index (BMI)

print("Kalkulator BMI")

berat = float(input("Masukkan berat badan (kg): "))
tinggi = float(input("Masukkan tinggi badan (m): "))

bmi = berat / (tinggi ** 2)
print(f"\nBMI kamu adalah: {bmi:.2f}")
 
if bmi < 18.5:
    print("Kategori: Kurus")
elif bmi < 25:
    print("Kategori: Normal")
elif bmi < 30:
    print("Kategori: Kelebihan berat badan")
else:
    print("Kategori: Obesitas")
```

```python
# Latihan 048 â€“ Sistem Nilai Mahasiswa
    
print("Sistem Penilaian Mahasiswa")
nilai = []

for i in range(1, 6):
    skor = float(input(f"Masukkan nilai mata kuliah {i}: "))
    nilai.append(skor)
 
rata = sum(nilai) / len(nilai)

if rata >= 90:
    grade = "A"
elif rata >= 80:
    grade = "B"
elif rata >= 70:
    grade = "C"
elif rata >= 60:
    grade = "D"
else:
    grade = "E"
 
print(f"\nRata-rata nilai: {rata:.2f}")
print(f"Grade akhir: {grade}")
```

```python
# Latihan 049 â€“ Permainan Tebak Angka dengan Nyawa

import random
  
rahasia = random.randint(1, 10)
nyawa = 3

print("ðŸŽ¯ Tebak angka antara 1â€“10 (Punya 3 kesempatan!)")

while nyawa > 0:
    tebakan = int(input("Masukkan tebakanmu: "))
    if tebakan == rahasia:
        print("ðŸŽ‰ Benar! Kamu menang!")
        break
    elif tebakan > rahasia:
        print("Terlalu tinggi!")
    else:
        print("Terlalu rendah!")
    nyawa -= 1
 
if nyawa == 0:
    print(f"ðŸ˜¢ Kesempatan habis! Angka rahasia adalah {rahasia}.")
```

```python
# Latihan 050 â€“ Kalkulator Bunga Sederhana

print("Kalkulator Bunga Sederhana")

modal = float(input("Masukkan modal awal (Rp): "))
bunga = float(input("Masukkan persen bunga (%): "))
waktu = float(input("Masukkan waktu (tahun): "))
  
hasil = (modal * bunga * waktu) / 100
total = modal + hasil

print(f"\nBunga yang diperoleh: Rp {hasil:,.2f}")
print(f"Total akhir tabungan: Rp {total:,.2f}")
```
